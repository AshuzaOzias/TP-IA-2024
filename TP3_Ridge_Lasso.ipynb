# Implémentation avec Scikit-learn
Voici comment implanter Ridge et LASSO avec Scikit-learn et comparer leurs performances sur le dataset Iris.



```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import joblib

#Charger et préparer le dataset Iris

iris = load_iris()
X = iris.data
y = iris.target

# Diviser le dataset en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Implémenter Ridge et LASSO
# Créer et entraîner le modèle Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

# Créer et entraîner le modèle LASSO
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

# Comparer les performances
# Prédictions
y_pred_ridge = ridge.predict(X_test)
y_pred_lasso = lasso.predict(X_test)

y_pred_ridge 
y_pred_lasso 

# Calculer l'erreur quadratique moyenne
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)

print(f'Ridge MSE: {mse_ridge}')
print(f'LASSO MSE: {mse_lasso}')

#Sauvegarder les modèles avec Joblib

joblib.dump(ridge, 'ridge_model.pkl')
joblib.dump(lasso, 'lasso_model.pkl')

# Visualiser les résultats
# Visualiser les coefficients
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(range(len(ridge.coef_)), ridge.coef_)
plt.title('Ridge Coefficients')

plt.subplot(1, 2, 2)
plt.bar(range(len(lasso.coef_)), lasso.coef_)
plt.title('LASSO Coefficients')

plt.show()
```
